```python
# Dependencies 
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
import pandas as pd 
```

```python
browser = Browser("chrome", executable_path="chromedriver", headless=True)
```

### WineBusiness.com

```python
# URL of wine-seacher
url = 'https://www.winebusiness.com/news'
browser.visit(url)
```

```python
# HTML object
html = browser.html
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(html, 'html.parser')
```

```python
# Return title, and preview

news_title = soup.find('div', class_='wb-section-item-title').text
news_p = soup.find('p', class_='wb-break-word').text

to_link = soup.find('div', class_='wb-section-item-title')
link = to_link.a['href']

print(f' Article Title: {news_title}')
print(f' Article Summary: {news_p}')
print('--------------------------')
print(f' Article Link: https://www.winebusiness.com{link}')
```

## Wine Folly Learn about Wines 

```python
# URL Wine Follow
url = 'https://winefolly.com/wines/'
browser.visit(url)
```

```python
# HTML object
html = browser.html
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(html, 'html.parser')
```

```python
# Find image url 
dict = {}
resources = soup.find_all('div', class_='col-md-4 col-sm-6 grape')
```

```python
for resource in resources:
    wine_link = resource.a['href']
    
    wine = resource('img')[1]
    wine_image = wine['src']
    
     # Update dictionary with new information
    dict.update({wine_image:wine_link})
    
    print('Adding to dictionary')
    print(wine_image)
    print(wine_link)
    print('-------')
```
